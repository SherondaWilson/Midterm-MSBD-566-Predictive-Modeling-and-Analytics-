```python
# Deep EDA for the uploaded Measles dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from textwrap import dedent
from IPython.display import display

# Helper function to display dataframes with a title
def display_dataframe_to_user(title, dataframe):
    """Displays a dataframe with a title."""
    print(f"--- {title} ---")
    display(dataframe)
    print("\n") # Add a newline for better separation

# -----------------------------
# 1) Load and initial cleanup
# -----------------------------
FILE_PATH = "/content/measles - measles.csv"
df_raw = pd.read_csv(FILE_PATH, low_memory=False)

# Keep a working copy
df = df_raw.copy()

# Standardize column names (strip/lower)
df.columns = [c.strip().lower() for c in df.columns]

# Ensure expected columns exist
expected_cols = [
    "state","year","name","type","city","county","enroll",
    "mmr","overall","xrel","xmed","xper","lat","lng"
]
missing_expected = [c for c in expected_cols if c not in df.columns]
meta_notes = []
if missing_expected:
    meta_notes.append(f"Missing expected columns (will skip related analyses): {missing_expected}")

# -----------------------------
# 2) Data Cleaning
# -----------------------------
# Convert numeric columns safely
for col in ["enroll","mmr","overall","xmed","xper","lat","lng"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Replace placeholder invalid values (-1) with NaN for rates/exemptions
for col in ["mmr","overall","xmed","xper"]:
    if col in df.columns:
        df.loc[df[col] < 0, col] = np.nan

# Clip rates to [0, 100] (some outliers like 169% are clearly invalid)
for col in ["mmr","overall","xmed","xper"]:
    if col in df.columns:
        df.loc[df[col] > 100, col] = np.nan  # set out-of-range to NaN
        df.loc[df[col] < 0, col] = np.nan

# Basic geofilter: USA-ish bounding box
if "lat" in df.columns and "lng" in df.columns:
    geo_mask = (
        df["lat"].between(18, 72, inclusive="both") &
        df["lng"].between(-180, -60, inclusive="both")
    )
    df.loc[~geo_mask, ["lat","lng"]] = np.nan

# Normalize year labels
if "year" in df.columns:
    df["year"] = df["year"].astype(str).str.strip()
    # Unify common variants
    df["year"] = df["year"].replace({
        "2017-201818": "2017-2018",
        "2017–2018": "2017-2018",
        "2018–2019": "2018-2019"
    })

# Herd immunity threshold indicator (>=95%)
if "mmr" in df.columns:
    df["herd_immune"] = df["mmr"] >= 95

# -----------------------------
# 3) High-level Summary
# -----------------------------
summary = {
    "n_rows_raw": len(df_raw),
    "n_rows_clean": len(df),
    "n_states": df["state"].nunique() if "state" in df.columns else np.nan,
    "n_schools": df["name"].nunique() if "name" in df.columns else np.nan,
    "years_present": sorted([y for y in df["year"].dropna().unique()]) if "year" in df.columns else []
}
summary_df = pd.DataFrame([summary])
display_dataframe_to_user("High-level summary", summary_df)

# Missingness report
missing_report = df[expected_cols].isna().mean().sort_values(ascending=False).rename("missing_fraction").to_frame() \
    if all(c in df.columns for c in expected_cols) else df.isna().mean().sort_values(ascending=False).rename("missing_fraction").to_frame()
display_dataframe_to_user("Missingness by column", missing_report.reset_index().rename(columns={"index":"column"}))

# -----------------------------
# 4) Descriptive Statistics
# -----------------------------
num_cols = [c for c in ["mmr","overall","xmed","xper","enroll"] if c in df.columns]
desc = df[num_cols].describe().T
display_dataframe_to_user("Descriptive statistics for key numeric columns", desc.reset_index().rename(columns={"index":"metric"}))

# -----------------------------
# 5) State-level Aggregations
# -----------------------------
if "state" in df.columns and "mmr" in df.columns:
    state_agg = (
        df.groupby("state")
          .agg(
              schools=("name","nunique"),
              records=("name","size"),
              median_mmr=("mmr","median"),
              mean_mmr=("mmr","mean"),
              pct_below_95=("mmr", lambda s: (s < 95).mean()*100),
              pct_below_90=("mmr", lambda s: (s < 90).mean()*100),
              median_overall=("overall","median")
          )
          .sort_values("median_mmr")
    )
    display_dataframe_to_user("State-level vaccination summary (sorted by median MMR ascending)", state_agg.reset_index())

# -----------------------------
# 6) Type-level Aggregations
# -----------------------------
if "type" in df.columns and "mmr" in df.columns:
    type_agg = (
        df.groupby("type", dropna=False)
          .agg(
              schools=("name","nunique"),
              records=("name","size"),
              median_mmr=("mmr","median"),
              mean_mmr=("mmr","mean"),
              pct_below_95=("mmr", lambda s: (s < 95).mean()*100)
          )
          .sort_values("median_mmr")
    )
    display_dataframe_to_user("School-type vaccination summary", type_agg.reset_index())

# -----------------------------
# 7) Top/Bottom Schools (by MMR)
# -----------------------------
if "mmr" in df.columns:
    # Avoid spurious small enrollments by filtering to enroll >= 20 when available
    filt = (df["enroll"] >= 20) if "enroll" in df.columns else pd.Series([True]*len(df), index=df.index)
    cols_for_view = [c for c in ["state","year","name","type","city","county","enroll","mmr","overall","xmed","xper"] if c in df.columns]
    top_schools = df[filt].dropna(subset=["mmr"]).sort_values("mmr", ascending=False).head(50)[cols_for_view]
    bottom_schools = df[filt].dropna(subset=["mmr"]).sort_values("mmr", ascending=True).head(50)[cols_for_view]
    display_dataframe_to_user("Top 50 schools by MMR (enrollment ≥ 20)", top_schools)
    display_dataframe_to_user("Bottom 50 schools by MMR (enrollment ≥ 20)", bottom_schools)

# -----------------------------
# 8) Correlations
# -----------------------------
if len(num_cols) >= 2:
    corr = df[num_cols].corr()
    display_dataframe_to_user("Correlation matrix (numeric features)", corr.reset_index().rename(columns={"index":"feature"}))

# -----------------------------
# 9) Visualizations (Matplotlib only; one chart per figure; no custom colors)
# -----------------------------

# Helper for safe figure creation
def safe_save(fig, path):
    fig.tight_layout()
    fig.savefig(path, dpi=150, bbox_inches="tight")

PLOT_DIR = "/content"
plots = []

# 9.1 Distribution of MMR
if "mmr" in df.columns:
    fig = plt.figure()
    plt.hist(df["mmr"].dropna(), bins=40)
    plt.title("Distribution of MMR vaccination rates (%)")
    plt.xlabel("MMR (%)")
    plt.ylabel("Count of schools")
    path = f"{PLOT_DIR}/plot_mmr_hist.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# 9.2 Distribution of Overall vaccination
if "overall" in df.columns:
    fig = plt.figure()
    plt.hist(df["overall"].dropna(), bins=40)
    plt.title("Distribution of Overall vaccination rates (%)")
    plt.xlabel("Overall (%)")
    plt.ylabel("Count of schools")
    path = f"{PLOT_DIR}/plot_overall_hist.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# 9.3 Boxplot of MMR by state (top 12 states by record count)
if "state" in df.columns and "mmr" in df.columns:
    top_states = df["state"].value_counts().head(12).index.tolist()
    data_box = [df.loc[df["state"] == s, "mmr"].dropna() for s in top_states]
    fig = plt.figure(figsize=(10,6))
    plt.boxplot(data_box, tick_labels=top_states, vert=True, showfliers=False)
    plt.title("MMR by state (top 12 states by record count)")
    plt.xlabel("State")
    plt.ylabel("MMR (%)")
    path = f"{PLOT_DIR}/plot_mmr_box_by_state.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# 9.4 Enrollment vs MMR scatter
if "enroll" in df.columns and "mmr" in df.columns:
    fig = plt.figure()
    plt.scatter(df["enroll"], df["mmr"], alpha=0.3)
    plt.title("Enrollment vs MMR")
    plt.xlabel("Enrollment")
    plt.ylabel("MMR (%)")
    path = f"{PLOT_DIR}/plot_enroll_vs_mmr.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# 9.5 Correlation heatmap
if len(num_cols) >= 2:
    fig = plt.figure()
    cax = plt.imshow(corr.values, aspect="auto")
    plt.title("Correlation heatmap")
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
    plt.yticks(range(len(corr.index)), corr.index)
    plt.colorbar(cax)
    path = f"{PLOT_DIR}/plot_corr_heatmap.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# 9.6 Geographic scatter (if lat/lng/ mmr available)
if all(c in df.columns for c in ["lat","lng","mmr"]):
    fig = plt.figure(figsize=(10,6))
    sc = plt.scatter(df["lng"], df["lat"], c=df["mmr"], s=5, alpha=0.5)
    plt.title("Geographic distribution of schools colored by MMR")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.colorbar(sc, label="MMR (%)")
    path = f"{PLOT_DIR}/plot_geo_mmr_scatter.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# 9.7 Personal exemptions vs MMR scatter
if all(c in df.columns for c in ["xper","mmr"]):
    fig = plt.figure()
    plt.scatter(df["xper"], df["mmr"], alpha=0.4)
    plt.title("Personal exemptions vs MMR")
    plt.xlabel("Personal exemptions (%)")
    plt.ylabel("MMR (%)")
    path = f"{PLOT_DIR}/plot_xper_vs_mmr.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# 9.8 Medical exemptions vs MMR scatter
if all(c in df.columns for c in ["xmed","mmr"]):
    fig = plt.figure()
    plt.scatter(df["xmed"], df["mmr"], alpha=0.4)
    plt.title("Medical exemptions vs MMR")
    plt.xlabel("Medical exemptions (%)")
    plt.ylabel("MMR (%)")
    path = f"{PLOT_DIR}/plot_xmed_vs_mmr.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# 9.9 Bar chart of states with lowest median MMR (bottom 10)
if "state" in df.columns and "mmr" in df.columns:
    bottom10 = (
        df.groupby("state")["mmr"]
          .median()
          .sort_values()
          .head(10)
    )
    fig = plt.figure()
    plt.bar(bottom10.index, bottom10.values)
    plt.title("States with lowest median MMR (bottom 10)")
    plt.ylabel("Median MMR (%)")
    plt.xticks(rotation=45, ha="right")
    path = f"{PLOT_DIR}/plot_bottom10_states_median_mmr.png"
    safe_save(fig, path)
    plots.append(path)
    plt.show()

# -----------------------------
# 10) Save cleaned data & summaries
# -----------------------------
CLEAN_PATH = "/content/measles_clean.csv"
df.to_csv(CLEAN_PATH, index=False)

STATE_SUMMARY_PATH = "/content/measles_state_summary.csv"
if "state" in df.columns and "mmr" in df.columns:
    state_agg.to_csv(STATE_SUMMARY_PATH)

NUM_SUMMARY_PATH = "/content/measles_numeric_summary.csv"
desc.to_csv(NUM_SUMMARY_PATH)

# If user notebook is present, just record that we followed a comparable flow
NOTE_PATH = "/content/eda_readme.txt"
with open(NOTE_PATH, "w") as f:
    f.write(dedent(f"""
    Deep EDA steps executed:
    - Loaded and cleaned data (replaced invalid rates, clipped to [0,100], filtered geocoords).
    - Generated missingness report and numeric summaries.
    - Built state/type aggregations and top/bottom school tables.
    - Plotted distributions, boxplots by state, scatter plots for relationships, correlation heatmap, and geoscatter.
    - Saved cleaned data and summaries.
    Additional notes: {('; '.join(meta_notes)) if meta_notes else 'None'}
    """))

# Return paths for convenience
{
    "cleaned_csv": CLEAN_PATH,
    "state_summary_csv": STATE_SUMMARY_PATH if "state_agg" in locals() else None,
    "numeric_summary_csv": NUM_SUMMARY_PATH,
    "note": NOTE_PATH,
    "plots": plots[:]
}
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report,
    roc_curve, auc
)
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the cleaned dataset
df = pd.read_csv("measles_clean.csv")

# 2. Data cleaning and preparation
# Drop invalid or missing vaccination rate entries
df = df[pd.to_numeric(df["mmr"], errors="coerce").notna()]
df["mmr"] = df["mmr"].astype(float)

# Define target: herd immunity achieved (1) or not (0)
df["herd_immune"] = (df["mmr"] >= 95).astype(int)

# Select relevant predictive features
features = ["overall", "xmed", "xper", "enroll", "lat", "lng"]
X = df[features].apply(pd.to_numeric, errors="coerce").fillna(0)
y = df["herd_immune"]

# 3. Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 4. Train Random Forest model
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    class_weight="balanced_subsample"
)
rf.fit(X_train, y_train)

# 5. Make predictions
y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]

# 6. Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
roc_auc = auc(*roc_curve(y_test, y_proba)[:2])

print("Model Performance Summary")
print(f"Accuracy: {accuracy:.3f}")
print(f"ROC-AUC: {roc_auc:.3f}")
print("\nConfusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# 7. Plot ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color="green", lw=2, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], color="gray", linestyle="--", label="Random Guessing")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve – Random Forest Model (Herd Immunity Prediction)")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# 8. Feature Importance Visualization
importances = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=True)
plt.figure(figsize=(8, 5))
sns.barplot(x=importances, y=importances.index, palette="viridis")
plt.title("Feature Importance in Random Forest Model")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.tight_layout()
plt.show()

# Optional: Export feature importance as table
importance_df = importances.reset_index()
importance_df.columns = ["Feature", "Importance"]
print("\nFeature Importance Table:")
print(importance_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split

# Load the dataset
df = pd.read_csv("/content/measles_clean.csv", low_memory=False)

# Clean and prepare data
df = df[pd.to_numeric(df["mmr"], errors="coerce").notna()]
df["mmr"] = df["mmr"].astype(float)

# Create target variable (1 = herd immunity, 0 = below)
df["herd_immune"] = (df["mmr"] >= 95).astype(int)

# Select predictive features (numerical + relevant)
features = ["overall", "xmed", "xper", "enroll", "lat", "lng"]
X = df[features].apply(pd.to_numeric, errors="coerce").fillna(0)
y = df["herd_immune"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Train Random Forest classifier
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)

# Predict probabilities for ROC curve
y_pred_proba = rf.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='green', lw=2, label=f'Random Forest ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve – Random Forest Model for Herd Immunity Prediction')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

roc_auc
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("/content/measles_clean.csv", low_memory=False)

# Ensure numeric types
for col in ["mmr", "overall", "xper", "xmed", "enroll"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Define helper metrics
def pct_below_95(s):
    s = pd.to_numeric(s, errors="coerce")
    return (s < 95).mean() * 100

# Aggregate by state
metrics = {
    "median_mmr": ("mmr", "median"),
    "median_overall": ("overall", "median"),
    "mean_xper": ("xper", "mean"),
    "mean_xmed": ("xmed", "mean"),
    "pct_schools_below95": ("mmr", pct_below_95),
    "n_records": ("mmr", "size")
}

state_agg = df.groupby("state").agg(**metrics).sort_values("n_records", ascending=False)

# Select the top 25 states by record count for readability
state_agg_top = state_agg.head(25).drop(columns=["n_records"])

# Prepare matrix for heatmap (z-score normalize columns for comparability)
mat = state_agg_top.copy()
mat_z = (mat - mat.mean()) / mat.std()

# Plot heatmap using matplotlib only (no seaborn)
fig, ax = plt.subplots(figsize=(10, 8))

im = ax.imshow(mat_z.values, aspect="auto")
ax.set_title("Figure 4. Exploratory Heatmap: State-Level Vaccination & Exemption Metrics (z-scored)")

# Set ticks and labels
ax.set_yticks(range(len(mat_z.index)))
ax.set_yticklabels(mat_z.index)

ax.set_xticks(range(len(mat_z.columns)))
ax.set_xticklabels(mat_z.columns, rotation=45, ha="right")

# Annotate with actual (non-z) values for interpretability
for i in range(mat_z.shape[0]):
    for j in range(mat_z.shape[1]):
        val = mat.iloc[i, j]
        if isinstance(val, (int, float)) and not np.isnan(val):
            if mat.columns[j].startswith("pct") or mat.columns[j].startswith("mean") or mat.columns[j].startswith("median"):
                text_val = f"{val:.1f}"
            else:
                text_val = f"{val:.0f}"
            ax.text(j, i, text_val, va="center", ha="center", fontsize=7, color="black")

# Add colorbar
cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Z-score (relative to column mean)")

plt.tight_layout()
plt.show()
```
